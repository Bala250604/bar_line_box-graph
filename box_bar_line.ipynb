```python
import pandas as pd
ds = pd.read_csv("HistoricalPrices.csv")
ds.columns
ds = ds.head(5)
```

```python
ds = ds.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
import matplotlib.pyplot as plt

plt.plot(ds['Date'], ds['Open'], color='green', label='Open')
plt.plot(ds['Date'], ds['Close'], color='red', label='Close')
plt.xlabel('Date')
plt.ylabel('Opening and Closing Price')
plt.show()
```

```python
plt.bar(ds['Date'], ds['Close'])
plt.show()
```

```python

plt.boxplot(ds['Close'])
plt.xlabel("Price")
plt.ylabel("Close")
plt.show()
```

```python
plt.scatter(ds['Date'], ds['Close'], color = 'r')
plt.scatter(ds['Date'], ds['Open'], color = 'g')
plt.show()
```
